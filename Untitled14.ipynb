#UNI,BI AND MULTI VARIATE ANALYSIS OF ADHAR ENORLLMENT

```python
import pandas as pd
import numpy as np
```

```python
import pandas as pd
import numpy as np

# Load enrolment dataset
enrol = pd.read_csv("/content/enrol.csv")

# Standardize column names
enrol.columns = enrol.columns.str.strip().str.lower()

# Convert date column
enrol["date"] = pd.to_datetime(enrol["date"], errors="coerce")

# Create year_month column
enrol["year_month"] = enrol["date"].dt.to_period("M").dt.to_timestamp()

# Create total_enrolment feature
enrol["total_enrolment"] = (
    enrol["age_0_5"] +
    enrol["age_5_17"] +
    enrol["age_18_greater"]
)

# Verify
print(enrol.head())
print(enrol.columns)
```

```python
# Summary statistics
enrol[["age_0_5", "age_5_17", "age_18_greater", "total_enrolment"]].describe()
```

```python
# Distribution of total enrolment
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(enrol["total_enrolment"], bins=30)
plt.title("Distribution of Total Aadhaar Enrolment")
plt.xlabel("Total Enrolment")
plt.ylabel("Frequency")
plt.show()
```

```python
# Age-wise enrolment contribution
enrol[["age_0_5", "age_5_17", "age_18_greater"]].sum().plot(
    kind="bar", figsize=(6,4),
    title="Age-wise Aadhaar Enrolment Distribution"
)
plt.ylabel("Total Enrolments")
plt.show()
```

```python
#bivariate
monthly_enrol = enrol.groupby("year_month")["total_enrolment"].sum()

plt.figure(figsize=(8,4))
plt.plot(monthly_enrol.index, monthly_enrol.values)
plt.title("Monthly Aadhaar Enrolment Trend")
plt.xlabel("Month")
plt.ylabel("Total Enrolment")
plt.show()

```

```python
state_avg = (
    enrol.groupby("state")["total_enrolment"]
    .mean()
    .sort_values(ascending=False)
)

state_avg.head(10).plot(kind="bar", figsize=(8,4))
plt.title("Top 10 States by Average Enrolment")
plt.ylabel("Average Enrolment")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(enrol["age_18_greater"], enrol["age_5_17"])
plt.xlabel("Adult Enrolment (18+)")
plt.ylabel("Child Enrolment (517)")
plt.title("Adult vs Child Enrolment Relationship")
plt.show()
```

```python
#muktivariate
import seaborn as sns

corr = enrol[
    ["age_0_5", "age_5_17", "age_18_greater", "total_enrolment"]
].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Between Enrolment Age Groups")
plt.show()
```

```python
pivot = enrol.pivot_table(
    values="total_enrolment",
    index="state",
    columns="year_month",
    aggfunc="sum"
)

plt.figure(figsize=(10,6))
sns.heatmap(pivot.fillna(0), cmap="RdYlBu")
plt.title("State-wise Monthly Aadhaar Enrolment Intensity")
plt.show()
```

#Monthly Enrolment Trend

```python
import matplotlib.pyplot as plt

monthly_enrol = (
    enrol
    .groupby("year_month")["total_enrolment"]
    .sum()
    .reset_index()
)

plt.figure(figsize=(10,5))
plt.plot(monthly_enrol["year_month"], monthly_enrol["total_enrolment"], marker="o")
plt.title("Monthly Aadhaar Enrolment Trend")
plt.xlabel("Month")
plt.ylabel("Total Enrolment")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Age-wise Enrolment Distribution

```python
age_totals = enrol[["age_0_5", "age_5_17", "age_18_greater"]].sum()

plt.figure(figsize=(6,4))
age_totals.plot(kind="bar")
plt.title("Age-wise Aadhaar Enrolment Distribution")
plt.xlabel("Age Group")
plt.ylabel("Total Enrolments")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#Top States by Enrolment

```python
state_enrol = (
    enrol
    .groupby("state")["total_enrolment"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
state_enrol.head(10).plot(kind="bar")
plt.title("Top 10 States by Aadhaar Enrolment")
plt.xlabel("State")
plt.ylabel("Total Enrolment")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

#Heatmap: State  Time Enrolment Intensity


```python
import seaborn as sns

pivot = enrol.pivot_table(
    values="total_enrolment",
    index="state",
    columns="year_month",
    aggfunc="sum"
)

plt.figure(figsize=(15,8))
sns.heatmap(pivot.fillna(0), cmap="YlGnBu")
plt.title("State-wise Monthly Aadhaar Enrolment Intensity")
plt.xlabel("Month")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```
